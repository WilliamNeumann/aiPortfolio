```python
# Bill Neumann
# Assignment 2

import torch
import torchvision
import torchvision.transforms as T
from PIL import Image

# Display the original image of a flower
img = Image.open('flower.jpg')
img.show()
```

```python
# Resize the image and display it
img = img.resize((416, 416))
img.show()
```

```python
# Convert image to a torch tensor
imgTensor = T.ToTensor()(img)

# Set the values for the mean and standard deviation
MEAN = (0.485, 0.456, 0.406)
STD = (0.229, 0.224, 0.225)

# Create the transformation
transform = T.Normalize(mean = MEAN, std = STD)
```

```python
# Normalize the iamge and convert it from a tensor back to a PIL Image
normalized_imgTensor = transform(imgTensor)
normalized_img = T.ToPILImage()(normalized_imgTensor)
```

```python
# Display the normalized image for comparison
normalized_img.show()
```
